```python
#obtencion de la cantidad del bucle
lis_lis = []

for i in range(len(urls)):

  url = urls[i]#'https://www.coingecko.com/es/categories/tron-ecosystem'
  response = requests.get(url)
  html_content = response.content
  # Supongamos que "html_content" contiene el HTML de la p√°gina web que est√°s escrapeando

  soup = BeautifulSoup(html_content, 'html.parser')

  # 1. Encuentra el div con la clase "blabla"
  div_blabla = soup.find('div', class_='tw-mt-4')

  # 2. Busca el div con la clase "x" dentro de "div_blabla"
  if div_blabla:
    div_x = div_blabla.find('div', class_='tw-flex tw-flex-1 tw-items-center tw-justify-evenly gecko-pagination sm:tw-justify-between')

    # 3. Busca el nav con la clase especificada dentro de "div_x"
    if div_x:
      nav = div_x.find('nav', class_='tailwind-reset pagy_nav tw-inline-flex tw-items-center tw-space-x-1 tw-rounded-lg gecko-pagination-nav')

      # 4. Busca el span sin atributos ni clases dentro del nav
      if nav:
        spans2 = nav.find('span',  class_='tw-relative tw-inline-flex tw-items-center tw-rounded-lg tw-px-4 tw-py-1.5 tw-text-sm tw-font-semibold tw-bg-primary-100 tw-text-primary-600 dark:tw-bg-primary-400/10 dark:tw-text-primary-400')
        #print(f'span con class : {spans2.text}')
        #lis_lis = [{'nombre': urls[i], 'cantidad': spans2}]
        lis_lis.append({'nombre': nombres_chains[i], 'cantidad': spans2.text})
#######################################################
        spans = nav.find_all('span', attrs={}, class_=False)

        # 5. Para cada span encontrado, busca las etiquetas <a> dentro de ellos
        for span in spans:
          a_tags = span.find_all('a', class_='tw-cursor-pointer tw-relative tw-inline-flex tw-items-center tw-rounded-lg tw-px-4 tw-py-1.5 tw-text-sm tw-font-semibold !tw-text-gray-900 hover:tw-bg-gray-50 dark:!tw-text-moon-50 dark:hover:tw-bg-moon-700')

          for a_tag in a_tags:
            # Haz algo con cada enlace encontrado
            text_A = a_tag.text
            #print(f'span sin class : {text_A}')
            #lis_lis = [{'nombre': urls[i], 'cantidad': spans}]
            lis_lis.append({'nombre': nombres_chains[i], 'cantidad': text_A})
            #print(text_A[len(a_tags)-1])

```

```python
# Diccionario para rastrear la √∫ltima aparici√≥n de cada elemento
ultimas_apariciones = {}

# Iterar sobre la lista de diccionarios en orden inverso
for diccionario in reversed(lis_lis):
    nombre = diccionario['nombre']
    cantidad = diccionario['cantidad']
    if nombre not in ultimas_apariciones:
        # Si es la primera aparici√≥n, agregarla al diccionario
        ultimas_apariciones[nombre] = diccionario
    else:
        # Si ya existe una aparici√≥n, verificar si es m√°s reciente que la √∫ltima registrada
        if cantidad > ultimas_apariciones[nombre]['cantidad']:
            ultimas_apariciones[nombre] = diccionario

# Convertir el diccionario de √∫ltimas apariciones a una lista
ultimas_apariciones_lista = list(ultimas_apariciones.values())

# Imprimir la lista resultante
for diccionario in ultimas_apariciones_lista:
    print(diccionario)
```

```python
#NOMBRE Y PRECIO
lista_nombres = []
lista_precios = []

#############################
##############################
blockchain = 'bitcoin' #chains minuscula
pagina = 1 #cantidad de paginas

#########################
##################################
#https://www.coingecko.com/es/categories/bitcoin-ecosystem
url = f'https://www.coingecko.com/es/categories/{blockchain}-ecosystem'
list_dic = []

for i in range(pagina):#cantidad de paginas
  count = 0
  response = requests.get(url)

  html_content = response.content


  soup = BeautifulSoup(html_content, 'html.parser')


  tbody = soup.find('tbody')



  if tbody:
    trs = tbody.find_all('tr')
    for tr in trs:

      div_td = tr.find('td', class_='tw-text-gray-900 dark:tw-text-moon-50 tw-px-1 tw-py-2.5 2lg:tw-p-2.5 tw-bg-inherit tw-sticky tw-left-[62px] md:tw-left-[72px]')
      if div_td:
        div = div_td.find('div')

        nombre = div.text.strip()

        nombre = nombre.replace('\n', '-')

        #lista_nombres.append(nombre)
        #print(f'nombre: {nombre}')
        #Tether--USDT
        partes = nombre.split('--')

        #if partes:
        #print(f'nombre: {partes[0]}\n')
        #print(f'token: {partes[1]}')

        #partes = cadena.split('--')
        #if len(partes) > 1:
        #lista_token.append(partes[1])



        span_td = tr.find('td', class_='tw-text-gray-900 dark:tw-text-moon-50 tw-px-1 tw-py-2.5 2lg:tw-p-2.5 tw-bg-inherit tw-text-end')
        print(span_td)
        if span_td:
          precio = span_td.find('span', attrs={
          'data-coin-id': True,
          'data-price-btc': True,
          'data-price-target': True})

          if precio is None:
            count += 1
            #lista_precios.append(None)
            #print(f'PRECIO: {None}')
            #print(f'chain: {blockchains[i]}\n')
            #print(f'ranking: {count}\n')
            token = partes[1] if len(partes) >= 2 else None
            list_dic.append( {'fecha': date.today(), 'blockchain': blockchain, 'token': token, 'nombre': partes[0],'precio': None, 'ranking': count} )
          else:
            count += 1
            #lista_precios.append(precio.text)
            #print(f'PRECIO: {precio.text}')
            #print(f'chain: {blockchains[i]}\n')
            #print(f'ranking: {count}\n')
            token = partes[1] if len(partes) >= 2 else None
            list_dic.append( { 'fecha': date.today(), 'blockchain': blockchain, 'token': token, 'nombre': partes[0], 'precio': precio.text, 'ranking': count})
```

```python
#FUNCION QUE GUARDA LOS DATOS PREVIAMENTE OBTENIDOS
def escribir_csv(datos, archivo):
    encabezados = ['fecha', 'blockchain', 'token', 'nombre', 'precio', 'ranking']
    with open(archivo, 'w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=encabezados)
        writer.writeheader()
        for dato in datos:
            writer.writerow(dato)
```

```python
# GUARDA LOS DATOS USANDO LA FUNCION ANTERIOR
archivo_csv = f'datos_{blockchain}.csv'
escribir_csv(list_dic, archivo_csv)
print(f"Los datos se han guardado en el archivo '{archivo_csv}'.")
```
